```python
import pandas as pd
```

```python
df = pd.read_csv('500_Cities__Local_Data_for_Better_Health__2019_release.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
df['MeasureId'].value_counts()
```

```python
sleepy_df = df[df['MeasureId']=='SLEEP'].copy()
```

```python
sleepy_df.head()
```

```python
sleepy_df.columns
```

```python
sleepy_df['Data_Value'].value_counts()
```

```python
round(sleepy_df['PopulationCount'] * (sleepy_df['Data_Value'] * .01), 0) #preview
```

```python
sleepy_df['percent_of_pop'] = round(sleepy_df['PopulationCount'] * (sleepy_df['Data_Value'] * .01), 0) #convert to percentage
```

```python
sleepy_df.groupby('CityName')['PopulationCount'].sum() #preview
```

```python
tot_pop_dict = dict(sleepy_df.groupby('CityName')['PopulationCount'].sum())
sleepy_df['tot_city_pop'] = sleepy_df['CityName'].map(tot_pop_dict)
```

```python
#preview
sleepy_df[['CityName','percent_of_pop', 'tot_city_pop']]
```

```python
perc_pop_dict = dict(sleepy_df.groupby('CityName')['percent_of_pop'].sum())
sleepy_df['perc_city_pop'] = sleepy_df['CityName'].map(perc_pop_dict)
```

```python
#preview
sleepy_df[['CityName','percent_of_pop', 'tot_city_pop', 'perc_city_pop']]
```

```python
#total number of people in the city divided by total city population = full city percentage
sleepy_df['tot_city_percent'] = round(sleepy_df['perc_city_pop'] / sleepy_df['tot_city_pop'],2)
```

```python
#preview
sleepy_df[['CityName','percent_of_pop', 'tot_city_pop', 'perc_city_pop', 'tot_city_percent']]
```

```python
sleepy_df.shape
```

```python
sleepy_df.drop_duplicates('CityName', inplace = True) #we only need one entry for each city now
```

```python
sleepy_df.shape
```

```python
sleepy_df.columns
```

```python
sleepy_final = sleepy_df[['Year', 'StateAbbr', 'StateDesc', 'CityName', 'Category', 'CategoryID',
                         'Short_Question_Text', 'tot_city_pop', 'perc_city_pop',
                         'tot_city_percent']].copy()
```

```python
sleepy_final.reset_index(inplace=True)
```
